# MATCH_SUMMARY

```python
import pandas as pd
```

```python
df = pd.read_html("https://www.espncricinfo.com/records/tournament/team-match-results/indian-premier-league-2023-15129")
```

```python
d = pd.DataFrame(df[0])
```

```python
d.columns
```

```python
d.shape
```

```python
d
```

```python
Match_ID = []
for i in range(0,len(d)):
    s=str(i)
    if i<10:
        p = "#230"
        ID = p+s
        Match_ID.append(ID)
    else:
        p1 = "#23"
        ID1 = p1+s
        Match_ID.append(ID1)
d["Match_ID"] = Match_ID
```

```python
d = d.drop("Scorecard", axis=1)
```

```python
d["Team 1"].unique()
```

```python
d["Team 2"].unique()
```

```python
d["Winner"].unique()
```

```python
d[d["Winner"]=="no result"]
```

```python
for j in range(len(d)):
    if d["Winner"][j]=="no result":
        d = d.drop(j, axis=0)
```

```python
d.head(45)
```

```python
d["Team 1"].unique()
```

```python
d.replace("Titans","Gujarat Titans", inplace=True)
d.replace("Super Giants","Lucknow Super Giants", inplace=True)
d.replace("Sunrisers","Sunrisers Hyderabad", inplace=True)
d.replace("RCB","Royal Challengers Bangalore", inplace=True)
d.replace("Super Kings","Chennai Super Kings", inplace=True)
d.replace("Capitals","Delhi Capitals", inplace=True)
d.replace("Royals","Rajastan Royals", inplace=True)
d.replace("KKR","Kolkata Knight Riders", inplace=True)
d.replace("Mumbai","Mumbai Indians", inplace=True)
```

```python
d.head()
```

```python
d.to_csv("Match_Summary.csv")
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
def battingstyle(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        p = soup.find_all('p')
        for i in range(len(p)):
            if p[i].text == "Batting Style":
                BattingStyle.append(p[i+1].text)
                break
        else:
            BattingStyle.append("NA")       
                
def bowlingstyle(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        pb = soup.find_all('p')
        for j in range(len(pb)):
            if pb[j].text == "Bowling Style":
                BowlingStyle.append(pb[j+1].text)
                break
        else:
            BowlingStyle.append("NA")
            
def playingrole(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        pr = soup.find_all('p')
        for k in range(len(pr)):
            if pr[k].text == "Playing Role":
                PlayingRole.append(pr[k+1].text)
                break
        else:
            PlayingRole.append("NA")        
```

```python
import requests
from bs4 import BeautifulSoup
def scrape_webpage(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        headings = soup.find_all('h1')
        p = soup.find_all('p')
        #a = soup.find_all('a')
        for heading in headings:
            PlayerName.append(heading.text)
        battingstyle(url)
        bowlingstyle(url)
        playingrole(url)
    else:
        print("Error:", response.status_code)
```
